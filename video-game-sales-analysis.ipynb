```python
a_data.drop(["Rank","Year"], axis=1, inplace=True)
a_data
```

```python
for nama in a_data["Name"].unique():
    print(nama)
```

```python
def fix_name(a_str):
    new_str = a_str.split('(')
    if len(new_str) == 1:
        return new_str[0]
    else:
        return new_str[0][:-1]

a_data['Name'] = a_data['Name'].apply(fix_name)
```
