```python
wine_data = pd.read_csv("https://raw.githubusercontent.com/benvictoria21/pandas/master/dataset/winequality-red.csv")
```

```python
X=wine_data.drop('quality',1) #feautres/independent variable
y=wine_data['quality'] #dependent /target variable

import statsmodels.api as sm
Xc=sm.add_constant(X)

model=sm.OLS(y,Xc).fit()
model.summary()
```
